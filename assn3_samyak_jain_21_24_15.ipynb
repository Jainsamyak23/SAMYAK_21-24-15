Q.5 Drop all rows with N/A,NA,na values in Num_Bedrooms

```python
df['NUM_BEDROOMS']=df['NUM_BEDROOMS'].replace('na',np.nan)
df=df.dropna(subset=['NUM_BEDROOMS'])
df
```

Q.6 Replace the NaN and String entries in Num_Bath with previous entry

```python
df['NUM_BATH']=df['NUM_BATH'].replace('HURLEY',np.nan)
df['NUM_BATH'].fillna(method='ffill',inplace=True)
df
```

Q.7 Replace the empty values in “SQ_FT” with the mean of the all entries.

```python
df['SQ_FT'] = df['SQ_FT'].replace('--', np.nan) #to replace string to Nan string
df['SQ_FT'] = pd.to_numeric(df['SQ_FT']) # to convert string values to float
mean_val = df['SQ_FT'].mean()
df['SQ_FT'].fillna(value=mean_val, inplace=True)
df
```
